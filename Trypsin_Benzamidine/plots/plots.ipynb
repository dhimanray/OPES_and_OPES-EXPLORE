```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate
```

```python
b30_40_data1 = np.loadtxt('Z_mixed_explore_opes_noextrabias_b30_40/convergence/deltaf.dat')
b30_40_data2 = np.loadtxt('Z_mixed_explore_opes_noextrabias_b30_40_trial_2/convergence/deltaf.dat')
b30_40_data3 = np.loadtxt('Z_mixed_explore_opes_noextrabias_b30_40_trial_3/convergence/deltaf.dat')

avg = np.mean(np.array([b30_40_data1[:40],b30_40_data2,b30_40_data3]),axis=0)
std = np.std(np.array([b30_40_data1[:40],b30_40_data2,b30_40_data3]),axis=0)

plt.plot(b30_40_data3[:,0],avg[:,2],c='green')
plt.fill_between(b30_40_data3[:,0],avg[:,2]+std[:,2],avg[:,2]-std[:,2],color='xkcd:lightgreen')
exp_values = -6.36*np.ones(len(b30_40_data3))
plt.plot(b30_40_data3[:,0],exp_values,c='k',lw=2,label='Experimental $\Delta G_b$')
plt.plot(b30_40_data3[:,0],exp_values-1,c='k',ls='--',label='$\pm$ 1 kcal/mol')
plt.plot(b30_40_data3[:,0],exp_values+1,c='k',ls='--')
plt.title('$\Delta E_{OPES-EXPLORE}$ = 30 kJ/mol; $\Delta E_{OPES}$ = 40 kJ/mol',fontsize=14)
plt.xlabel('Simulation Time (ns)',fontsize=16)
plt.ylabel('$\Delta G_b$ (kcal/mol)',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(-12,0)
plt.legend(fontsize=14)
plt.savefig('ex_30_opes_40.png',dpi=200,bbox_inches='tight')
```
